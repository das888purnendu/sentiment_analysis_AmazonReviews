```python
from selenium import webdriver
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.set_page_load_timeout(20)
```

```python
driver.get("https://www.amazon.in/Rockerz-255-Bluetooth-Immersive-Black/product-reviews/B07C2VJXP4/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews")
```

```python
rev=[] #main database of the reviews
for i in driver.find_elements_by_css_selector('span.a-icon-alt,.review-text-content span'):
    line = i.get_attribute('innerText')
    try:
        rating = float(line[0:3])
    except ValueError:
        rev.append([rating,line])
    
print("Total : ",len(rev),"Reviews Added")
print("Page 1 Done !")
```

```python
counter=1
while driver.find_element_by_class_name("a-last"):
    time.sleep(2) # If internet speed is high. else make it 10
    driver.find_element_by_class_name("a-last").click()
    counter+=1
    l1=len(rev)
    for i in driver.find_elements_by_css_selector('span.a-icon-alt,.review-text-content span'):
        line = i.get_attribute('innerText')
        try:
            rating = float(line[0:3])
        except ValueError:
            rev.append([rating,line])
    print("Total : ",(len(rev)-l1),"Reviews Added")
    print("Page",counter," Done !")
```

```python
driver.close()
```
